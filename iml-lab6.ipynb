```python
#kukadiya janki n
#216270307088
#Practical No.6: Write a Pandas program to implement following operation
#to create the mean and standard deviation of the data of a given series

import pandas as pd

s=pd.Series([12,45,78,23,56,89,88])

print("series is:\n",s)
print("\nmean of s:",s.mean())
print("\nstandard deviation of s:",s.std())
```
