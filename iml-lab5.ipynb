```python
#kukadiya janki n
#216270307088
#Practical No.5: Write a NumPy program to implement following operation
#to find the maximum and minimum value of a given flattened array

import numpy as np

arr=np.array([10,28,88,50,55,75,46])

minimum=np.amin(arr)
print("minimum value is :",minimum)

maximum=np.amax(arr)
print("maximum value is :",maximum)
```

```python
#kukadiya janki n
#216270307088
#Practical No.5: Write a NumPy program to implement following operation
#to compute the mean, standard deviation, and variance of a given array along the second axis

import numpy as np

arr=np.array([[10,20,30],[40,50,60]])
print("orignal array is:",arr)

mean=np.mean(arr,axis=1)
print("mean is:",mean)

std_dev=np.std(arr,axis=1)
print("standard deviation is:",std_dev)

veriance=np.var(arr,axis=1)
print("veriance is:",veriance)
```
